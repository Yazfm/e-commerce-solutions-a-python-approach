#### Libraries used

```python
import pandas as pd
import numpy as np
```

# Read the csv file

```python
e_commerce_df = pd.read_csv('../datasets/e_commerce_raw.csv', encoding='mac_roman')
```

# Clean the dataset

```python
e_commerce_df.head()
```

```python
e_commerce_df.info()
```

We see that the shape of the dataset is ~542K records with 8 columns.

#### Rename columns

```python
e_commerce_df.columns = ['invoice_no', 'stock_code', 'description', 'quantity', 'invoice_date', 'unit_price', 'customer_id', 'country']
```

```python
e_commerce_df.head()
```

#### Check null values

```python
e_commerce_df.isna().sum()
```

#### Drop null value rows

```python
e_commerce_df = e_commerce_df.dropna().reset_index(drop=True)
```

```python
e_commerce_df.shape
```

```python
e_commerce_df.info()
```

#### Data types

```python
e_commerce_df.dtypes
```

```python
e_commerce_df.head()
```

```python
# Change invoice_date tot datetime
e_commerce_df.invoice_date = pd.to_datetime(e_commerce_df.invoice_date)
```

```python
e_commerce_df.head()
```

```python
# Change customer_id to string
# Get rid of trailing '0' from float by converting the type to 'int' before 'str'
e_commerce_df.customer_id = e_commerce_df.customer_id.astype(int).astype(str)
```

```python
e_commerce_df.describe()
```

#### Negative quantity?

```python
# 'quantity' has negative values?

e_commerce_df[e_commerce_df.quantity<0]
```

```python
# Drop negative quantity values as we don't have enough information to why they are negative
# Use the inverse '~' of any records with negative quantity

e_commerce_df = e_commerce_df[~e_commerce_df.quantity<0].reset_index(drop=True)
```

```python
e_commerce_df
```

#### Unit price = 0 ?

```python
e_commerce_df[e_commerce_df.unit_price==0].head()
```

```python
# No clear indicator of why unit_price = 0
# Drop these records

e_commerce_df = e_commerce_df[e_commerce_df.unit_price>0].reset_index(drop=True)
```

```python
# Export clean dataframe for EDA
e_commerce_df.to_csv('../datasets/e_commerce_clean_non_null.csv')
```
